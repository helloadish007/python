### STACK IN PYTHON 

A stack is a LIFO (Last In First Out — the element placed at last can be accessed at first) structure which can be commonly found in many programming languages.
This structure is named as “stack” because it resembles a real-world stack — a stack of plates.

```python
#STACK IN PYTHON 
class Stack():
    def __init__(self):
        self.list=[]
    def push(self,item):
        self.list.append(item)
    def pop(self):
        self.list.pop()
    def TOS(self):
        if not self.is_empty():            #TOS:TOP OF STACK
            return self.list[-1]
        else:
            s='stack empty!'
            return s
    def display(self):
        return self.list
    def is_empty(self):
        return self.list==[]
s1=Stack()
s1.push('adish')
s1.push('aneesh')
s1.push('asish')
print(s1.display())
print(s1.TOS())
s1.pop()
print(s1.display())
```

### BALANCED PARANTHESIS IN ORDER: (APPLICATION OF STACK DATA STRUCTURE)

```python
def balanced_paranthesis(s):
    o="({["
    c=")}]"
    matching={")":"(","}":"{","]":"["}
    stack=[]
    for i in s:
        if i in o:
            stack.append(i)
        elif i in c:
            if len(stack)==0:
                return False
            elif stack[-1]==matching[i]:
                stack.pop()
            else:
                return False
    return len(stack)==0
```

# Three Number Sum:

Write a function that takes in a non-empty array of distinct integers and an integer representing a target sum.The function should find all triplets in the array that sum up to the target sum and return a 2D array of all these triplets.The numbers in each triplet should be ordered in ascending order and the triplets themselves should be ordered in ascending order with respect to the numbers they hold.

if no 3 numbers sum up to the target sum,the function should return an empty array.

### input:

array=[12,3,1,2,-6,5,-8,6]
target sum=0

### output:

[[-8,2,6],[-8,3,5],[-6,1,5]]



```python
def threenumbersum(a,target):
    a.sort()
    lt=[]
    for i in range(len(a)):
        l=i+1
        r=len(a)-1
        while l<r:
            curr_sum=a[i]+a[l]+a[r]
            if curr_sum==target:
                lt.append([a[i],a[l],a[r]])
                l+=1
                r-=1
            elif curr_sum<target:
                l+=1
            elif curr_sum>target:
                r-=1
    return lt
threenumbersum([12,3,1,2,-6,5,-8,6],0 )
```
